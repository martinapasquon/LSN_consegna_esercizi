L'obiettivo di questa esercitazione è di scrivere un codice che utilizzi la tencnica del Simulated Annealig  per 
ottenere una stima del ground state di una particella quantistica in 1D soggetta al potenziale:
    $$
V(x) = x^4 - \frac{5}{2}x^2
$$
    
La funzione d'onda di prova utilizzata  in questo caso è
$$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$
e il codice deve calcolare il valore di aspettazione dell'Hamiltoniana
$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$
Per fare questo innanzitutto si scelgono dei valori per la temperatura T, $\mu$
e $\sigma$ e  si utilizza l'algoritmo di Metropolis per campionare $|\Psi_T^{\sigma\mu}(x)|^2$.
Si stima quindi l'Hamiltoniana $H_{old} $ valutando $\frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}$ nei valori campionati e si trova 
il valor medio con il metodo a blocchi. Quindi si
propongono dei nuovi valori per $\mu$ e $\sigma$ ($\mu_{new}$ e $\sigma_{new}$)
e calcola l'hamiltoniana corrispondente $H_{new} $. Si pone  $\mu = \mu_{new}$ e $\sigma=\sigma_{new}$ con probabilità 
P=min[1, $ e^{-\frac{(H_{new}-H_{old})}{T}}$]. Questo processo viene ripetuto per $n_T$ diversi valori di temperatura,
ognuno minore del precedente e per ogni T si effettuano k diverse proposte di  $\mu_{new}$ e $\sigma_{new}$.
I parametri utlizzati in questa simulazione sono:
    
- numero di blocchi N=10;

- numero di step per ogni blocco=10;

- valori iniziali di $\mu$ e $\sigma$ $\mu=0.6$ $\sigma=0.6$;

- k=15;

- numero di step del SA $n_T=500$;

- numero di valori campionati da $|\Psi_T^{\sigma\mu}(x)|^2$ per ogni step  $n_x=100$.

```python


def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

print("Ground state energy: ", E[0])
print("1st excited state energy: ", E[1])
print("2nd excited state energy: ", E[2])

```

```python

def psi2_analitic(x, mu , sigma):
    f1 = np.e**((-(x-mu)**2)/(2*(sigma**2)))
    f2 = np.exp((-(x+mu)**2)/(2*(sigma**2)))
    return ((f1+f2)**2)/(2*np.sqrt(2*np.pi)*sigma)# +np.exp(-(mu*sigma)**2))

```

```python
n_bins=100
mu=0.570422
sigma= 0.822685
plt.figure(figsize=(8,5))
x = np.linspace(-a/2, a/2, N)
psi2_sort= np.loadtxt("C:es8/output_psi2.txt", unpack='true')
plt.hist(psi2_sort, n_bins, density=True, facecolor='g', range=(-3, 3))
plt.plot(x,(psi[0])**2, label="soluzione numerica")
plt.plot(x,psi2_analitic(x, mu, sigma), label="curva analitica", color="black")

plt.title("Potential  Probabilities")
plt.xlabel("x")
plt.grid(True)
plt.xlim((-2,2,))
plt.legend()
plt.ylim((0.,1))
plt.show()

```
