### Esercizio 3.1
Questo esercizio si propone di calcolare al tempo $t=0$ il prezzo di call-option Europeo, 
$C[S(0),0]$, e quello di put-option,$P[S(0),0]$. 
I parametri utilizzati in questo esercizio sono:
- prezzo del bene a $t=0$: $S(0)=100$
- tempo nel futuro in cui viene comprato il bene: $T=1$
- prezzo prestabilito: $K=100$
- tasso d'interesse: $r=0.1$
- volatilità: $\sigma=0.25$.
    
    
$C[S(0),0]$ e $P[S(0),0]$ in questo esercizio vengono calcolati in due diversi modi:
1. campionando il prezzo del bene finale $S(T)$</span> for a $GBM(r,\sigma^2)$
2. campionando l'andamento discreto $GBM(r,\sigma^2)$ del prezzo del bene 
dividendo $[0,T]$ in $10^2$ intervalli di tempo: $S(0)$, $S(t_1)$, $S(t_2)$, ... fino a  $S(t_{100}=T=1)$.

Per quanto riguarda il primo caso, $S\sim GBM(\mu,\sigma^2)$, con valore iniziale $S(0)$, si ha
    $$S(t) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right]$$
dove $W(t)\sim\mathcal{N}(0,t)$.
Nel secondo caso invece, poichè $W$ sono indipendenti e distributi su una gaussiana, l'andamento discreto $S\sim GBM(\mu,\sigma^2)$ per 
$t_1 < t_2 < … < t_n$ viene campionato con la seguente formula ricorsiva:
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
    con $Z_i \sim \mathcal{N}(0,1)$

#### caso 1:
Il grafico sottostante rappresenta i valori di $C[S(0),0]$ e $P[S(0),0]$ ottenuti nella simuazione in funzione 
del numero di blocchi. In questo esercizio sono stati utlizzati M=100000 step divisi in N=100 blocchi.In questo
grafico e nel successivo sono rappresentati anche i valori di $C[S(0),0]$ e di $P[S(0),0]$ ottenuti analiticamente. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

M=100000
N=100                 
L=int(M/N)            
x = np.arange(N)

call= 14.975790778311286
put=  5.4595325819072364

plt.rcParams.update({'font.size': 17})
plt.figure(figsize=(10,5)) 

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

C_dir, err_C_dir = np.loadtxt("C:es03.1/C_dir.txt", usecols=(0,1), delimiter=' ', unpack='true')
plt.errorbar(x,C_dir ,yerr=err_C_dir, label="C ")
plt.axhline(y=call, color='black', linestyle='--')
plt.axhline(y=put, color='black', linestyle='--')
P_dir, err_P_dir = np.loadtxt("C:es03.1/P_dir.txt", usecols=(0,1), delimiter=' ', unpack='true')
plt.errorbar(x,P_dir ,yerr=err_P_dir, label="P ")

plt.xlabel('throws')
plt.ylabel('price')
plt.title('P and C by sampling directly the final asset price for a GMB')
plt.grid(True)
plt.legend()

plt.show()
```
