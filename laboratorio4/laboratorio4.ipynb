### Esercizio  4.02: 
Questo esercizio si propone di effettuare una simulazione di dinamica molecolare attraverso il modello di Lennard-Jones 
dell'Argon nello stato solido, liquido e gassoso e di utilizzare il metodo a bocchi per trovare i valori medi progressivi 
dell'energia totale interna per particella (E/N), dell'energia potenziale per particella(U/N), dell'energia cinetica per particella
(K/N),
della temperatura (T) e della pressione (P) in unità ridotte. In particolare, le tre diverse condizioni in cui si effettua la
simulazione sono:
1. fase solida: $\rho^\star = 1.1$, $T^\star = 0.8$ (raggio di cut-off : $r_c = 2.2$)
2. fase liquida: $\rho^\star = 0.8$, $T^\star = 1.1$ (raggio di cut-off: $r_c = 2.5$)
3. fase gassosa: $\rho^\star = 0.05$, $T^\star = 1.2$ (raggio di cut-off: $r_c = 5.0$)
    
    
Per effettuare la simulazione innanzitutto si deve raggiungere l'equilibrio 
in ognuno dei tre stati e per farlo si parte dalla configurazione cristallina (così da evitare di avere molecole sovrapposte 
nella configurazione iniziale) a una temperatura T tale che, all'equilibrio, 
il sistema raggiunga la temperatura T*. Il grafico sottostante rappresenta il valor medio della temperatura dalla condizione
iniziale al raggiungimento dell'equilibrio, nel caso solido, liquido e gassoso. Il grafico è stato ottenuto con N=2000
blocchi da uno step nel caso liquido e solido, mentre è stato ottenuto con N=4000 blocchi da uno step nel caso gassoso, poichè
tale stato raggiunge più lentamente degli altri due l'equilibrio.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
#T=1.55
plt.figure(figsize=(7,3))            
plt.rcParams.update({'font.size': 15})    
n_step=2000
x = np.arange(n_step)
x1 = np.arange(2*n_step)
#T_solid=np.zeros(n_step)
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

solid, errs= np.loadtxt("C:MDNVE_MCNVT/output_temp_oscill_solid.txt", usecols=(2,3), unpack='true')
plt.errorbar(x,solid ,yerr=errs, label="solido")
plt.xlabel('numero di step')
plt.ylabel('T')
plt.title('oscillazioni della temperatura')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(7,3))
liquid, errl= np.loadtxt("C:MDNVE_MCNVT/output_temp_oscill_liquid.txt", usecols=(2,3), unpack='true')
plt.errorbar(x,liquid ,yerr=errl, label="liquido")
plt.xlabel('numero di step')
plt.ylabel('T')
plt.title('oscillazioni della temperatura')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(7,3))
gas, errg= np.loadtxt("C:MDNVE_MCNVT/output_temp_oscill_gas.txt", usecols=(2,3), unpack='true')
plt.errorbar(x1,gas ,yerr=errg, label="gassoso")
plt.xlabel('numero di step')
plt.ylabel('T')
plt.title('oscillazioni della temperatura')
plt.legend()
plt.grid(True)

plt.show()
```

Dal grafico sopra si evince che la temperatura non conserva il suo valore iniziale. Nel caso liquido e solido
questo potrebbe essere dovuto al fatto che 
in questo processo si passa dalla condizione in cui il potenziale è minimo (stato solido) e si fa aumentare man mano.
In questo modo
la forza utlizzata nell'alogritmo di verlet (-grad(U)) fa diminuire man mano le velocità e di conseguenza la temperatura.

Dopo aver raggiunto l'equilibrio si effettua la simulazione ponendo come condizione iniziale quella del sistema dopo che ha raggiunto
l'equilibrio e si procede al calcolo del valor medio progressivo dell'energia potenziale per particella,
dell'energia cinetica per particella, dell'energia totale interna per
particella, della temperatura,  e della pressione.  L'energia potenziale
totale (U) viene calcolata come $$U= 4 \sum_{i=1, j>i}^{N}(\frac{1}{|dr_{ij}|^{12}} +\frac{1}{|dr_{ij}|^{6}}) $$
dove la sommatoria è sul numero di molecole e $ dr_ {ij} $ indica la distanza tra le coppie di molecole con condizioni
al contorno periodiche. La formula è quella del potenziale di Lennard-Jonesin in unità naturali. Il grafico sottostante 
rappresenta il valor medio progressivo dell'energia potenziale per particella (V/N) negli stati solido, liquido e gassoso.

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

x = np.arange(50)
plt.figure(figsize=(10,10))
solidu, errsu= np.loadtxt("C:MDNVE_MCNVT/output_epot_solid.txt", usecols=(2,3), unpack='true')
plt.errorbar(x,solidu ,yerr=errsu, label="solido")
liquidu, errlu= np.loadtxt("C:MDNVE_MCNVT/output_epot_liquid.txt", usecols=(2,3), unpack='true')
plt.errorbar(x,liquidu ,yerr=errlu, label="liquido")
gasu, errgu= np.loadtxt("C:MDNVE_MCNVT/output_epot_gas.txt", usecols=(2,3), unpack='true')
plt.errorbar(x,gasu ,yerr=errgu, label="gassoso")

plt.xlabel('numero di blocchi')
plt.ylabel('U/N')
plt.title("valor medio progressivo dell'energia potenziale per particella")
plt.legend()
plt.grid(True)

plt.show()
```

L'energia cinetica totale (K) viene calcolata come $$ K= \sum_{i=1}^{N} \frac{1}{2}m v_i^{2} $$ 
dove $v_i$ e m sono rispettivamente la velocità e la massa delle singole molecole. Il grafico sottostante 
rappresenta il valor medio progressivo dell'energia cinetica per particella (K/N) negli stati solido, liquido e gassoso.

```python
plt.figure(figsize=(10,10))

solidK, errsK= np.loadtxt("MDNVE_MCNVT/output_ekin_solid.txt", usecols=(2,3), unpack='true')
plt.errorbar(x,solidK ,yerr= errsK, label="solido")
liquidK,  errlK= np.loadtxt("C:MDNVE_MCNVT/output_ekin_liquid.txt", usecols=(2,3), unpack='true')
plt.errorbar(x,liquidK ,yerr= errlK, label="liquido")
gasK,  errgK= np.loadtxt("C:MDNVE_MCNVT/output_ekin_gas.txt", usecols=(2,3), unpack='true')
plt.errorbar(x,gasK ,yerr= errgK, label="gassoso")

plt.xlabel('numero di blocchi')
plt.ylabel(' K/N')
plt.title("valor medio progressivo dell'energia cinetica per particella")
plt.legend()
plt.grid(True)

plt.show()

```

L'energia totale interna (E) viene calcolata come $$ E= K + U $$. Il grafico sottostante 
rappresenta il valor medio progressivo dell'energia totale interna per particella (E/N) negli stati solido, liquido e gassoso.

```python
plt.figure(figsize=(10,10))
solidEtot, errsEtot= np.loadtxt("C:MDNVE_MCNVT/output_etot_solid.txt", usecols=(2,3), unpack='true')
plt.errorbar(x,solidEtot ,yerr= errsEtot, label="solido")
liquidEtot,  errlEtot= np.loadtxt("C:MDNVE_MCNVT/output_etot_liquid.txt", usecols=(2,3), unpack='true')
plt.errorbar(x,liquidEtot ,yerr= errlEtot, label="liquido")
gasEtot,  errgEtot= np.loadtxt("C:MDNVE_MCNVT/output_etot_gas.txt", usecols=(2,3), unpack='true')
plt.errorbar(x,gasEtot ,yerr= errgEtot, label="gassoso")

plt.xlabel('numero di blocchi')
plt.ylabel('E/N')
plt.title(" valor medio progressivo dell'energia totale interna per particella")
plt.legend()
plt.grid(True)

plt.show()
```

La temperatura (T) viene calcolata come $$ T= \frac{2}{3} K/N.$$Il grafico sottostante 
rappresenta il valor medio progressivo della temperatura negli stati solido, liquido e gassoso.

```python
plt.figure(figsize=(10,10))
x = np.arange(50)
solidT, errsT= np.loadtxt("C:MDNVE_MCNVT/output_temp_solid.txt", usecols=(2,3), unpack='true')
plt.errorbar(x,solidT ,yerr=errsT, label="solid")
liquidT, errlT= np.loadtxt("C:MDNVE_MCNVT/output_temp_liquid.txt", usecols=(2,3), unpack='true')
plt.errorbar(x,liquidT ,yerr=errlT, label="liquid")
gasT, errgT= np.loadtxt("C:MDNVE_MCNVT/output_temp_gas.txt", usecols=(2,3), unpack='true')
plt.errorbar(x,gasT ,yerr=errgT, label="gas")
plt.xlabel('numero di blocchi')
plt.ylabel('Temperatura ')
plt.title('valor medio progressivo della temperatura')
plt.legend()
plt.grid(True)

plt.show()
```

La pressione (P) viene calcolata come 
$$P= \rho T +48 \sum_{i=1, j>i}^{N}(\frac{1}{|dr_{ij}|^{12}} +\frac{1}{2|dr_{ij}|^{6}}) \frac{1}{3 V}$$
dove vol indica il volume del sistema e $\rho $ la densità.

Il grafico sottostante 
rappresenta il valor medio progressivo della pressione negli stati solido, liquido e gassoso.

```python
plt.figure(figsize=(10,10))

solidp, errsp= np.loadtxt("C:MDNVE_MCNVT/output_ptot_solid.txt", usecols=(2,3), unpack='true')
plt.errorbar(x,solidp ,yerr=errsp, label="solido")
liquidp, errlp= np.loadtxt("C:MDNVE_MCNVT/output_ptot_liquid.txt", usecols=(2,3), unpack='true')
plt.errorbar(x,liquidp ,yerr=errlp, label="liquido")
gasp, errgp= np.loadtxt("C:MDNVE_MCNVT/output_ptot_gas.txt", usecols=(2,3), unpack='true')
plt.errorbar(x,gasp ,yerr=errgp, label="gassoso")

plt.xlabel('numero di blocchi')
plt.ylabel("P")
plt.title("valor medio progressivo della pressione")
plt.legend()
plt.grid(True)

plt.show()
```
