```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

M=100000
N=100                 
L=int(M/N)            
x = np.arange(N) 
err_prog=np.zeros(N)
sum_prog=np.zeros(N)

file_sum = open("C:/Users/marti/lab1/laboratorio1/randomgen/sum_prog.txt"", "r")
i=0
for line in file_sum:
    sum_prog[i]=float(line)
    #print(sum_prog[i])
    i+=1
    
file_sum.close()
i=0
file_err = open("C:/Users/marti/lab1/laboratorio1/r/sum_prog2.txt"", "r")
for line in file_err:
    err_prog[i]=float(line)
    i+=1

file_err.close()  
x*=L # Number of throws = block * (Number of throws in each block)
plt.errorbar(x,sum_prog-0.5,yerr=err_prog)

plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```
