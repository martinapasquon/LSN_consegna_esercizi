### ESERCIZIO 01.1
#### punto1 
Questo esercizio si propone di stimare  $\langle r \rangle = \int_0^1 r dr = 1/2$. 
con la relativa incertezza utilizzando M numeri random estratti uniformemente dall'intervallo  [0,1]
e sfruttando il metodo a blocchi. Per fare ciò si dividono gli M numeri random  in N blocchi e 
 si calcola il valor medio progressivo e il rispettivo errore degli elementi di ogni blocco e di tutti i blocchi precedenti. 
Nel primo grafico è illustrata la media progressiva -0.5 con il rispettivo errore, entrambi in funzione di M.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import numpy as np
import scipy as sp
from scipy import stats
from scipy.stats import cauchy


M=100000
N=100                 
L=int(M/N)            
x = np.arange(N)

plt.rcParams.update({'font.size': 18})

plt.figure(figsize=(10,5))            


x*=L
sum_prog, err = np.loadtxt("C:es01.1/es1/sum_prog.txt", usecols=(0,1), delimiter=' ', unpack='true')
plt.errorbar(x,sum_prog-0.5 ,yerr=err)
plt.xlabel('numero di lanci')
plt.ylabel('$<r> - 1/2$')
plt.title('valor medio distribuzione uniforme')
plt.grid(True)

plt.show()
```

### ESERCIZIO 01.1
#### punto 3
Questo esercizio si propone di implementare il test chi-quadrato di Pearson per n numeri random estratti unformemente
dall'intervallo [0,1]. Per fare ciò si divide tale intevallo in M sotto intervalli identici, si generano n numeri casuali 
compresi tra 0 e 1 e, per ciascun sotto intervallo, si conta quanti di questi n numeri random sono all'interno del sotto 
intervallo considerato. Il chi- quadro in questo caso viene stimato come:
$$\chi^2 = \sum_{i=1}^M \frac{\left( n_i - n/M \right)^2}{n/M}$$

La somma è sul numero di sotto-intervalli M e $n_i $ è il numero di numeri random presenti nel singolo sotto intevallo. n/M indica 
sia il valor medio che la varianza. Infatti il valor medio della distribuzione binomiale è pari al numero di numeri random
generati (n) moltiplicato per la probabilità che il singolo numero random sia all'interno dell'i-esimo intervallo (1/M).
La varianza di una distribuzione binomiale invece è pari a np(1-p) ma in questo caso, poichè M è grande (quindi p è piccola), è 
stata approssimata come np. Questo test è stato ripetuto 100 volte e l'istogramma sottostante rappresenta i risultati ottenuti.
Considerato il valore della varianza e che i numeri random vengono estratti uniformemente, ci si aspetta in media che
$(n_i - n/M)^2 \simeq n/M$ e quindi che il chi-quadro abbia in media un valore pari a M. L'istogramma è centrato
su 100 e questo conferma quanto scritto sopra.

```python
n_bins = 50

plt.figure(figsize=(10,5))            

chi_quadro = np.loadtxt("C:es01.1/es3/chi_quadro.txt", unpack='true')
plt.hist(chi_quadro, n_bins, density=True, facecolor='g', range=(45, 200))
plt.ylabel('throws')
plt.xlabel('$\chi^{2}$')
plt.title('test del $\chi^{2}$')

plt.show()
```

### ESERCIZIO 01.2
Questo esercizio si propone di verificare il teorema del limite centrale. Per fare ciò vengono generati tre istogrammi
riempiti ognuno con n numeri casuali pari a $S_N = \frac{1}{N}\sum_{i=1}^N x_i$ (per $N=1, 2, 10, 100$), in cui $x_i$ è una 
variabile estratta in un primo caso da una distribuzione gaussiana con $\mu$=0 e $sigma$=1, poi da una distribuzione esponenziale 
con parametro $\lambda$=1 e infine da una distribuzione di Cauchy-Lorentz con parametri (0,1). L'istogramma per N=100 
contenente i numeri random estratti da una distribuzione espoenziale può essere fittato con una gaussiana con media=$\lambda $ e 
varianza= $ \frac{1}{\lambda N}$. La stessa cosa accade per tutti gli isogrammi contententi numeri random estratti dalla distribuzione
gaussiana mentre non accade per quelli contenenti numeri estratti dalla distribuzione di Cauchy-Lorentz. Tutto ciò è in accordo
con il teorema del limite centrale: infatti un' ipotesi di tale teorema è che i primi due momenti della distribuzione siano 
finiti e le distribuzioni normale ed esponenziale verificano tale ipotesi, mentre  il secondo momento di quella di 
Cauchy-Lorentz è infinito.

la gaussiana di media=0 e varianza=1/N.

### ESERCIZIO 01.3
Questo esercizio consiste in una simulazione dell'esperimento di Buffon. L'esperimento ha come scopo quello di ottenere una stima 
di $\pi$ e consiste nella caduta di un ago di lunghezza L su un foglio con righe parallele distanziate tra loro di d. La probabilità
che l'ago cada su una di tali righe è pari a $P = 2L/\pi d$. Se l'ago viene fatto cadere M volte e cade su una riga $N_{hit}$ 
volte allora si può ottenere una stima di $\pi$ da: 
  $$\pi = \frac{2L}{Pd} = \lim_{N_{thr} \to \infty}\frac{2LN_{thr}}{N_{hit}d}$$
Per simulare la caduta dell'ago, per ogni misura si generato due numeri random. Il primo si estrae da una distribuzione 
uniforme tra 0 e d e rappresenta la posizione del baricentro dell'ago, il secondo si estrae dall'intervallo (0, $\pi $) e 
rappresenta l'angolo $/theta$ tra la posizione dell'ago e quella delle righe. Per generare quest'ultimo numero senza utilizzare
$\pi$ si generano due numeri x e y estratti da una distribuzione uniforme, x appartenente a [-1,1]e y appartenente a [0,1] e si
genera $\theta$ come 
$$\theta=arccos(\frac{x}{\sqrt{x^2 +y^2}}) $$
Anche per questo esercizio è stato utilizzato il metodo a blocchi e il grafico sottostante rappresenta la media 
progressiva e l'errore progressivo sulla stima di $\pi$. 


```python
x = np.arange(N)  
M=10000
N=100
L=int(M/N)  
x*=L
plt.figure(figsize=(10,5)) 
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
buffon, err = np.loadtxt("C:es01.3/sum_prog.txt", usecols=(0,1), delimiter=' ', unpack='true')
plt.errorbar(x ,buffon, yerr=err)
plt.xlabel('numero di lanci')
plt.ylabel('$\pi$')
plt.title('simulazione esperimento Buffon')
plt.grid(True)

plt.show()
```
