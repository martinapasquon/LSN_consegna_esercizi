### Esercizio  02.1
Questo esercizio si propone di calcolare l'integrale 
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$
con il metodo Monte Carlo in due diversi modi. Nel primo caso si estraggono M numeri random x da una distribuzione
uniforme e nel secondo invece da una distribuzione non unifrome che, nell'intervallo di integrazione, 
ha un andamento simile alla funzione da integrare. In entrambi i casi si trovano i valori medi e i rispettivi
errori della stima di questo integrale con il metodo a blocchi (M lanci e N blocchi). Il calcolo dell'integrale nel primo caso
si effettua come $$I_{unif}=\frac{1}{n} \sum_{i=1}^{M}f(x_i) $$
dove  f indica l'integranda e $x_i$ sono numeri random estratti da una distribuzione di probabilià uniforme tra [0,1].
Il calcolo dell'integrale nel secondo caso si effettua come $$I_{IS}=\frac{1}{n}\sum_{j=1}^{M} \frac{f(x_j)p(x_j)}{q(x_j)} $$
dove  p indica la distribuzione uniforme, q quella non uniforme e $x_i$ sono numeri casuali tra [0,1] estratti dalla q.
La q scelta in questo esercizio è pari a $q(x)=-2x+2$ e di conseguenza $x_i=1-\sqrt{1-rand} $ (dove a sinistra dell'uguale
vi è l'inversa della cumulata di g e rand è un numero random tra 0 e 1 estratto dalla distribuzione uniforme).
Nel grafico sottostante sono rappresentati i valori medi con i rispettivi errori in funzione del numero di lanci M. Dalla 
lunghezza delle barre di errore e dai valori medi ottenuti si evince che -a parità di M- l'estrazione di x dalla
distribuzione non uniforme fornisce risultati più accurati di quella dalla distribuzione uniforme.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

M=100000
N=200                 
L=int(M/N)            
x = np.arange(N)
x*=L

plt.rcParams.update({'font.size': 17})
plt.figure(figsize=(10,5))            
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

sum, err = np.loadtxt("C:es02.1/int_prog.txt", usecols=(0,1), delimiter=' ', unpack='true')
plt.errorbar(x,sum ,yerr=err, label="da una distribuzione uniforme ")

sum1, err1 = np.loadtxt("C:es02.1/int_prog2.txt", usecols=(0,1), delimiter=' ', unpack='true')
plt.errorbar(x,sum1 ,yerr=err1,  label="utilizzando l' importance sampling")
plt.xlabel('M')
plt.ylabel('valor medio integrale')
plt.title('intergrale con metodo MC')
plt.legend(loc="lower right")
plt.grid(True)

plt.show()
```

### Esercizio  02.2
Questo esercizio si propone di simulare M random walks in tre dimensioni sia su un reticolo che nel continuo,
entrambi con passo a=1 e di trovare, per entrambe le tipologie di random walk, 
$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ al variare dello step i utlizzando il metodo a blocchi.
Per fare ciò si divide M in N blocchi e, per ogni step del RW,  si calcolano la media e il relativo errore progressivi. Il grafico 
sottostante rappresenta il valor medio e il relativo errore di $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ al variare dello
step e il fit con una fuzione $f(i) \approx \sqrt{i}$ (dove i indica lo step i-esimo). Dal grafico è evidente che entrambe le funzioni hanno lo stesso andamento 
di f(x) e questo ne conferma l'andamento diffusivo. In questo esercizio i RW sono costituiti da 100 passi, M=10000 e N=100.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit


N = 100
x = np.arange(N)
y=np.zeros(N)
def f(x,k):  # defining the fitting function
    return k * np.sqrt(x)

k=2 #8/np.sqrt(300)

plt.figure(figsize=(20,10))


plt.rc('text', usetex=True)
plt.rc('font', family='serif')


plt.subplot(1, 2, 1)
sum_c, err_c = np.loadtxt("C:es02.2/RW _continuum.txt", usecols=(0,1),  delimiter=' ', unpack='true')
plt.errorbar(x,sum_c ,yerr=err_c,color='orange', label="RW in the continuum")

p_opt_c, p_cov_c = curve_fit(f, x, sum_c)
y_fit = f(x,p_opt_c)
plt.plot(x,y_fit, label="fit continuum", color="black") # plotting fitted function

plt.xlabel('numero di blocchi')
plt.ylabel('$<R>$')
plt.title('RW 3D continuum')
plt.legend(loc="lower right")
plt.grid(True)


plt.subplot(1, 2, 2)
sum_l, err_l = np.loadtxt("C:es02.2/RW _lattice.txt", usecols=(0,1), delimiter=' ', unpack='true')
plt.errorbar(x,sum_l ,yerr=err_l, color="green",label="RW on a cubic lattice")
p_opt_l, p_cov_l = curve_fit(f, x,sum_l)
y_fit_l = f(x,p_opt_l)
plt.plot(x,y_fit_l, label="fit lattice ", color="red") # plotting fitted function

plt.xlabel('numero di blocchi')
plt.ylabel('$<R>$')
plt.title('RW 3D lattice')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
print("optimized parameters in the continuum k =")
print(p_opt_c)
print("parameters uncertainty =")
print(np.sqrt(np.diagonal(p_cov_c)))
print("optimized parameters on a cubic lattice k =")
print(p_opt_l)
print("parameters uncertainty =")
print(np.sqrt(np.diagonal(p_cov_l)))
```
