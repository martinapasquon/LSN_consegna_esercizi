### Esercizio 09.1

L'obiettivo di questo esercizio è quello di risolvere il problema del commesso viaggiatore utilizzando un algoritmo genetico.
Date N città, l'esercizio quindi consiste nel trovare il cammino più corto a condizione che tutte le città vengano visitate 
una e una sola volta e che alla fine del percorso si ritorni alla città da cui si parte.
Per risolvere questo problema si è deciso di rappresentare il percorso delle città con una matrice $N\times N$  in cui
ogni riga e ogni colonna hanno uno e un solo elemento uguale a 1 e tutti gli altri uguali a zero. In tal modo
l'i-esima colonna rappresenta l'i-esima città visitata e la j-esima riga di ogni colonna rappresenta la città visitata. 
Per svolgere l'esercizio innanzitutto si crea un operatore di commutazione (commmuta due città in un percorso) e di controllo
(verifica che i perocorsi soddisfano i vincoli) e con questi due operatori si genera una popolazione iniziale costituita 
da $M=5000$ diversi percorsi.  Quindi si  ordina la popolazione da quella corrispondente al percorso più breve 
a quella corrispondente al più lungo e si crea un operatore di selezione che seleziona l'individuo $j$ con l'algoritmo:
$j= int(M \times r^p)+1$ dove $r$ è un numero radom  $r\in[0,1)$ e $p=4$ in questo esercizio. 
In seguito si implementano gli operatori  di shift di n posizioni per m città contigue, di commutazione di n città 
contigue, di inversione dell'ordine in cui appaiono m città e di crossover tra due diverse città.                                                          
Le mutazioni e il crossover vengono applicati con probabilità rispettiavmente $P_m \approx 0.1$ e $P_c \approx 0.9$.
Gli individui ai quali si applica il crossover vengono scelti con l'operatore di selezione. Così facendo si ottengono
popolazioni con individui molto diversi tra loro e si agisce soprattutto sui percorsi più brevi, quindi questi ulitimi tendono a
migliorare al passare delle generazioni. In questa simuazione vengono generate $N_{gen=5000}$ diverse generazioni.
                                                           

#### Caso1: città su un cerchio
Il grafico sottostante rappresenta la media della lunghezza degli M/2 percorsi più corti e la lunghezza del percorso più breve 
in funzione del numero di generazioni nel caso in cui si hanno 34 città su un cerchio di raggio r=10.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
#grafico della Temperatura che oscilla, 5000 blocchi da 1. Per Solidi. T iniziale=0.7

plt.figure(figsize=(10,5))            
plt.rcParams.update({'font.size': 15})                
          
x = np.arange(5000)

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

bestavL= np.loadtxt("C:es9/L_best_av_out_circle.txt", unpack='true')
plt.plot(x,bestavL, label="average L")
bestL= np.loadtxt("C:es9/L_best_out_circle.txt", unpack='true')
plt.plot(x,bestL, label="best L")

plt.xlabel('numero di generazioni')
plt.ylabel('L')
plt.title('lunghezza percorso città su un cerchio di raggio r')
plt.legend()
plt.grid(True)

plt.show()
```

Il grafico sottostante rappresenta il percorso migliore in coordinate cartesiane.

```python
plt.figure(figsize=(5,5))
xcoord, ycoord= np.loadtxt("C:es9/best_cartesian_circle.txt", usecols=(0,1), unpack='true')
plt.plot(xcoord,ycoord, linestyle="solid")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Percorso migliore città su un cerchio')
plt.grid(True)

plt.show()
```

Il percorso migliore trovato rappresenta effettivamente il percorso più breve possibile e questo fatto conferma che 
il codice implementato funziona in modo adeguato.
