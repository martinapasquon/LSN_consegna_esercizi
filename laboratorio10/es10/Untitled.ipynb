```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
#grafico della Temperatura che oscilla, 5000 blocchi da 1. Per Solidi. T iniziale=0.7

             
          
x = np.arange(5000)

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

bestavL= np.loadtxt("C:L_best_av_out.txt", unpack='true')
plt.plot(x,bestavL, label="average L")
bestL= np.loadtxt("C:L_best_out.txt", unpack='true')
plt.plot(x,bestL, label="best L")

plt.xlabel('numero di generazioni')
plt.ylabel('L')
plt.title('lunghezza percorso città su un cerchio di raggio r')
plt.legend()
plt.grid(True)

plt.show()
```

```python
xcoord, ycoord= np.loadtxt("C:best_cartesian.txt", usecols=(0,1), unpack='true')
plt.plot(xcoord,ycoord, linestyle="solid")
plt.grid(True)

plt.show()
```
