### Esercizio 10.2
Questo esercizio si propone di utilizzare il codice dell'esercizio 9.01 per trovare il percorso migliore tra quelli che
passano una e una sola volta in ogni capitale di ogni stato americano. L'obiettivo finale è quello di verificare
quanto migliora la ricerca del percorso più corto al variare del numero di nodi  e di $N_{migr}$.
In questo caso si effettuano 4 diverse ricerche:
1. ricerca con 1 nodo;
2.ricerca con 2 nodi e $N_{migr}=30$;
3.ricerca con 3 nodi e $N_{migr}=30$;
4.ricerca con 3 nodi e senza migrazioni.
In ognuna delle quattro simulazioni si hanno M=5000 diversi percorsi per ogni individuo e $N_{gen}= 5000 $generazioni.


Il grafico sottostante rappresenta la media della lunghezza degli M/2 percorsi più corti e la lunghezza del percorso più breve 
in funzione del numero di generazioni per ognuna delle 4 simulazioni.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

        
x = np.arange(5000)



plt.rc('text', usetex=True)
plt.rc('font', family='serif')


plt.rcParams.update({'font.size': 17})
plt.figure(figsize=(20,20)) 

plt.subplot(2,2,1)
bestavL= np.loadtxt("C:es10\L_best_av_out_america_1.txt", unpack='true')
plt.plot(x,bestavL, label="average L")
bestL= np.loadtxt("C:es10\L_best_out_america_1.txt", unpack='true')
plt.plot(x,bestL, label="best L")

plt.xlabel('numero di generazioni')
plt.ylabel('L')
plt.title('lunghezza percorso, 1 nodo')
plt.legend()
plt.grid(True)

plt.subplot(2,2,2)
bestavL= np.loadtxt("C:es10\L_best_av_out_america_2.txt", unpack='true')
plt.plot(x,bestavL, label="average L")
bestL= np.loadtxt("C:es10\L_best_out_america_2.txt", unpack='true')
plt.plot(x,bestL, label="best L")

plt.xlabel('numero di generazioni')
plt.ylabel('L')
plt.title('lunghezza percorso, 2 nodi')
plt.legend()
plt.grid(True)

plt.subplot(2,2,3)
bestavL= np.loadtxt("C:es10\L_best_av_out_america_3.txt", unpack='true')
plt.plot(x,bestavL, label="average L")
bestL= np.loadtxt("C:es10\L_best_out_america_3.txt", unpack='true')
plt.plot(x,bestL, label="best L")

plt.xlabel('numero di generazioni')
plt.ylabel('L')
plt.title('lunghezza percorso, 3 nodi ')
plt.legend()
plt.grid(True)

plt.subplot(2,2,4)
bestavL= np.loadtxt("C:es10\L_best_av_out_america_3_noexc.txt", unpack='true')
plt.plot(x,bestavL, label="average L")
bestL= np.loadtxt("C:es10\L_best_out_america_3_noexc.txt", unpack='true')
plt.plot(x,bestL, label="best L")

plt.xlabel('numero di generazioni')
plt.ylabel('L')
plt.title('lunghezza percorso, 3 nodi senza migrazione')
plt.legend()
plt.grid(True)


plt.show()
```

Il grafico sopra conferma il fatto che la ricerca migliora all'aumentare del numero di nodi: infatti con N nodi 
si hanno N diversi processi in parallelo che cercano il percorso più breve e questo non può che migliorare il codice. 
Inoltre il grafico conferma anche che in generale la ricerca è migliore nel caso in cui si hanno migrazioni perchè la migrazione
è come una mutazione e aumenta la difersificazione tra gli individui di una popolazione.
