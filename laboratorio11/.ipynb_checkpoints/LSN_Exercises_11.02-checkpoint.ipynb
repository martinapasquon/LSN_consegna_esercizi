#  Esercizio 11.1
L'obiettivo di questo esercizio è di migliorare il fit della retta $$f(x)=2x+1$$ 
cambiando i valori del numero di  epoche, $N_{\mathrm{epochs}}$, 
di dati  $N_{\mathrm{train}}$ e il rumore $\sigma$.


Come prima cosa si crea una funzione che effettua la regressione e che prende in ingresso  $N_{\mathrm{epochs}}$, 
$N_{\mathrm{train}}$ e $\sigma$.

```python
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python


def model_function(n_train, sigma_i,n_epochs ):
    np.random.seed(0)
    # target parameters of f(x) = m*x + b
    m = 2 # slope
    b = 1 # intersect
    x_train = np.random.uniform(-1, 1, n_train)
    x_valid = np.random.uniform(-1, 1, 50)
    x_valid.sort()
    
    y_target = m * x_valid + b # ideal (target) linear function
    sigma = 0.4 # noise standard deviation
    
    y_train = np.random.normal(m * x_train + b, sigma_i) # actual measures from which we want to guess regression parameters
    y_valid = np.random.normal(m * x_valid + b, sigma_i)
    # compose the NN model
    model = tf.keras.Sequential()
    model.add(Dense(1, input_shape=(1,)))
    # compile the model choosing optimizer, loss and metrics objects
    model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
    history = model.fit(x=x_train, y=y_train,  batch_size=32, epochs=n_epochs,shuffle=True, validation_data=(x_valid, y_valid), verbose=0)
    model.get_weights()
    # evaluate model
    ev = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)
    # print performance
  

    # evaluate model with the exact curve
    score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

    # print performance
    print('n_train=' ,n_train, 'sigma=', sigma_i,'N epochs=', n_epochs)
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    # look into training history
    # Plot training & validation loss values
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model loss')
    plt.xlabel("N epochs")
    plt.ylabel("loss")
    plt.legend(['Train', 'Test'], loc='best')
    
    plt.show()
    x_predicted = np.random.uniform(-1, 1, 100)
    y_predicted = model.predict(x_predicted)
    plt.scatter(x_predicted, y_predicted,color='r')
    plt.title("x predicted ")
    plt.xlabel("x ")
    plt.ylabel("y ")
    plt.plot(x_valid, y_target)
    plt.grid(True)
    plt.show()
```

Si effettua quindi il fit per  diversi valori di  $N_{\mathrm{train}}$.In questo esercizio  
$N_{\mathrm{train}}$ va da 500 a 4000 con passo di 500.

```python
n_train_arr=np.arange(500, 4000, 500 )
epochs=20
sigma=0.2
for i in range(len(n_train_arr)):
    model_function(n_train_arr[i], sigma,epochs )
    
```

Dai dati si evince che il fit  migliora (ma sempre più lentamente) al crescere di $N_{\mathrm{train}}$.Per questo motivo si sceglie
$N_{\mathrm{train}}= 1500$. Poi si effettua il fit
per diversi valori di  $N_{\mathrm{epoch}}$. In questo esercizio  $N_{\mathrm{epoch}}$ va da 20 a 100 con passo di 20.


```python
n_train=1500
n_epochs_arr=np.arange(20, 100, 20 )
sigma=0.2
for i in range(len(n_epochs_arr)):
    model_function(n_train, sigma, n_epochs_arr[i])
```

Anche in questo caso il fit  migliora (ma sempre più lentamente) al crescere di $N_{\mathrm{epochs}}$ (anche se in generale
si rischia di effettuare un overfitting se si utilizza un valore  troppo alto di  $N_{\mathrm{epochs}}$).                                                                                                      .
Per questi motivi si sceglie
$N_{\mathrm{train}}= 1500$. Poi si effettua il fit
per diversi valori di  $N_{\mathrm{epoch}}$. In questo esercizio  $N_{\mathrm{epoch}}$ va da 20 a 100 con passo di 20.


Da i dati ottenuti si evince che il fit migliora al crescere di $N_{\mathrm{epoch}}$. Si pone quindi  $N_{\mathrm{epoch}}$=100
e si effettua il fit per diversi valori di  $\sigma$. In questo esercizio  $\sigma$ va da 0 a 4 con passo di 4.

```python
sigma_i_arr=np.arange(0., 4, 0.5)
epochs=100
for i in range(len(sigma_i_arr)):
    model_function(n_train, sigma_i_arr[i], epochs)
```

I dati ottenuti mostrano come il fit peggiori al crescere di $\sigma$.
